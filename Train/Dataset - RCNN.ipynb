```python
from torchvision import transforms, datasets, models
import torch
from torch import optim, cuda
from torch.utils.data import DataLoader, sampler, random_split
import torch.nn as nn

from PIL import Image
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
cv2.setUseOptimized(True);
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()

def getboxes(image):
    ss.setBaseImage(image)
    ss.switchToSelectiveSearchFast()
    imout = image.copy()
    return imout, ss.process()

def get_iou(bb1, rect):
    X,Y,X2,Y2 = bb1
    x,y,w,h = rect
    m = abs(X - x) + abs(Y - y) + abs(X2 - (x+w)) + abs(Y2 - (y+h))
    return m

def parse_one_annot(path):
    data = {}
    for i,e in enumerate(os.listdir(path)):
        try:
            data_bach = pd.read_csv(os.path.join("Label",e[:-4]+".csv"))
            data_bach = data_bach.values[0]
            image = cv2.imread(os.path.join("Image",e))
            x1 = int((data_bach[0].split(" ")[0]).split(".")[0])
            y1 = int((data_bach[0].split(" ")[1]).split(".")[0])
            x2 = int((data_bach[0].split(" ")[2]).split(".")[0])
            y2 = int((data_bach[0].split(" ")[3]).split(".")[0])
            data[e] = [x1,y1,x2,y2]
        except:
            pass
    return data

a = parse_one_annot("Image")
```

```python
counter = 0
for filename in list(a.keys()):
    try:
        istake = True
        file = os.path.join("Image",filename)
        image = cv2.imread(file)
        imout, boxes = getboxes(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        print(counter)
        for boxe in boxes[:1000]:
            iou = get_iou(a[filename], boxe)
            x,y,w,h = boxe
            if iou < 50:
                timage = imout[y:y+h,x:x+w]
                resized = cv2.resize(timage, (128,128), interpolation = cv2.INTER_AREA)
                cv2.imwrite("dataset/1/"+str(counter)+".jpg", cv2.cvtColor(resized, cv2.COLOR_RGB2BGR))
                counter += 1
            if iou > 300 and istake == True:
                istake = False
                timage = imout[y:y+h,x:x+w]
                resized = cv2.resize(timage, (128,128), interpolation = cv2.INTER_AREA)
                cv2.imwrite("dataset/0/"+str(counter)+".jpg", cv2.cvtColor(resized, cv2.COLOR_RGB2BGR))
                counter += 1
    except Exception as e:
        #print(e)
        print("error in "+filename)
        #continue
```
